###  Explanation of **kws (keyword arguments)** in FacetGrid

The `kws` dictionary in FacetGrid is used to pass **custom styling parameters** for the plot.  
In this case, the `kws` dictionary is controlling the appearance of **scatter points**.

---

| Argument            | What It Does                                                                 |
|--------------------|-------------------------------------------------------------------------------|
| **s=50**            | Sets the **size of the scatter points (dots)** to **50 pixels**.              |
| **linewidth=0.5**   | Adds a **thin border** around each scatter point with a **width of 0.5**.     |
| **edgecolor='black'** | Makes the **border color** of each scatter point **black**.                 |

---

###  Example Usage
```python
kws = dict(s=50, linewidth=0.5, edgecolor='black')
g = g.map(plt.scatter, 'CriticRating', 'AudienceRating', **kws)
